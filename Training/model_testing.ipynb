```python
import numpy as np

from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.xception import preprocess_input
```

```python
model=load_model("Breath-Scan.h5")
```

```python
img_path=r"C:\\Users\\nalla_enejoi3\\OneDrive\\Desktop\\My_Projects\\COVID-19_Radiography_Dataset\\Normal\\images\\Normal-1.png"
img = load_img(img_path, target_size=(299, 299))
img
```

```python
img_array = img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)
```

```python
pred=np.argmax(model.predict(img_array))
pred
```

```python
index=['Covid','Lung Opacity','Normal','Viral Pneumonia']
output=index[pred]
print("The person Diagnoised is : ",output)
```
